##Introduction on bitcoin
Bitcoin-USD (BTCUSD) is the spot trading pair that quotes the price of one bitcoin in U.S. dollars on cryptocurrency markets, with the price set by continuous supply and demand across exchanges rather than by any central authority. Bitcoin’s market price is influenced by its fixed supply schedule and halving events, on-chain transaction costs and throughput, macro sentiment and activity in derivatives like futures and perpetual swaps.

- BTC-USD is the quoted exchange rate between Bitcoin and U.S. dollars, commonly displayed as BTCUSD on crypto venues and data dashboards.
- Bitcoin is a decentralized digital currency that operates on a peer-to-peer network without a central issuer and the USD leg is simply the fiat pricing unit for the pair.

Transactions are signed with private keys and broadcast to a peer‑to‑peer network, where miners bundle them into blocks and compete via proof‑of‑work to add blocks to the chain, preventing double‑spending and establishing consensus. The blockchain is a distributed public ledger replicated by nodes across the network rather than maintained by a single institution. This design allows two parties to transact directly online with settlement recorded transparently on the ledger.

Bitcoin’s issuance follows fixed rules: new coins are minted as block rewards that halve roughly every four years, with a maximum supply capped at 21 million BTC. Each bitcoin is divisible into 100 million satoshis, enabling very small payments and fractional ownership. Because supply is inelastic and predetermined, market price is driven by demand in open markets rather than by monetary policy.
- Demand responds to macro conditions, risk appetite, narratives and adoption trends, so price changes are driven by evolving expectations as well as the deterministic supply path.

### Trading mechanics
- On centralized exchanges, order books match buy and sell orders and the last traded price becomes the market price, which is determined by supply and demand rather than a central bank or administrator.
- Aggregate market prices reflect cross-exchange trading and arbitrage, with the pair notation BTCUSD providing a common reference for analysis tools and data feeds.

### Volatility and risks
- Bitcoin’s price can be volatile because the asset class is comparatively smaller than major global markets since as already mentioned above tsupply is inelastic, so marginal demand shifts can move price materially.
- Some economists characterize bitcoin as speculative or bubble-like, while others view it as a store-of-value competitor to gold; risk assessments vary widely across reputable institutions.
-For that reason, it could be necessary to construct a regime-dependent analysis in order to extract the possible market regimes.

### On-chain vs market price
- The network processes transactions in blocks roughly every 10 minutes, with transactions queued in a mempool and prioritized by fees, but the on-chain settlement layer does not set the BTC-USD price.
- Fees and throughput can affect user experience and timing of flows, indirectly influencing trading behavior and short-term liquidity conditions observed in BTCUSD markets.

###Time series exploratory analysis

A **unit root process** behaves like:

$$
y_t = y_{t-1} + u_t
$$

where $u_t$ is white noise.

* If the process has a unit root ($\rho = 1$), shocks **persist forever** → nonstationary.
* If $|\rho| < 1$, shocks die out → stationary.

The ADF test is used to decide between these two.

It is based on estimating:

$$
\Delta y_t = \alpha + \beta t + \gamma y_{t-1} + \sum_{i=1}^p \delta_i \Delta y_{t-i} + \epsilon_t
$$

where:

* $\Delta y_t = y_t - y_{t-1}$
* $\alpha$ = drift (optional)
* $\beta t$ = trend (optional)
* $\sum \delta_i \Delta y_{t-i}$ = extra lags to “whiten” residuals

and the hypothesis are:

* $H_0$: unit root → nonstationary ($\gamma = 0$)
* $H_1$: stationary ($\gamma < 0$)

The **ADF test statistic** is the **t-statistic** for $\gamma$:

$$
ADF = \frac{\hat{\gamma}}{\text{SE}(\hat{\gamma})}
$$

* If it’s very **negative**, we reject the unit root hypothesis.
* If it’s close to 0, we can’t reject unit root.

In addition:

* If **p-value < 0.05** → reject $H_0$. Series is stationary.
* If **p-value > 0.05** → fail to reject $H_0$. Series has a unit root (nonstationary).


As we can see, the returns are obviously stationary because we reject the $H_0$.

#Step 2: Parkinson estimator as estimator of the volatility

Since one of our goal is constructing the regimes for the bitcoin returns, another important feature for the hidden Markov models is the volatility.

In this case, I would like to use the Parkinson estimator for assessing the vol.

The **Parkinson estimator** (1980) is a range-based estimator of volatility.
Unlike the simple variance of returns, it makes use of **intraday high and low prices**, which contain more information about price variability than just closing prices.
In the specific, for a given asset, over a sample of $T$ trading days, the Parkinson volatility estimator is:

$$
\hat{\sigma}_{P}^2 = \frac{1}{4 \ln(2) \, T} \sum_{t=1}^{T} \left( \ln \frac{H_t}{L_t} \right)^2
$$

Where:

* $H_t$: the high price on day $t$
* $L_t$: the low price on day $t$
* $\ln \frac{H_t}{L_t}$: the log range of the asset on that day

The estimator of volatility (standard deviation) is then:

$$
\hat{\sigma}_{P} = \sqrt{\hat{\sigma}_{P}^2}
$$

The reasons behind my choice are:
* the daily **range** (high–low) reflects the **true variability** of prices within the day;
* using this information gives a **more efficient estimator** of volatility compared to the classical close-to-close estimator (about **5 times more efficient** under a Brownian motion assumption).
* however, it **ignores opening and closing prices** and **cannot handle jumps or drift well**, but since it is going to be used for the regime analysis with the returns calculated with close prices, it could be important to have an additional measure.



```python
def parkinson_estimator(prices):
  high_low_return=np.log(bitcoin["High"]/bitcoin["Low"])
  hv_daily=np.sqrt((1/(4*np.log(2)*len(bitcoin["High"]))))*high_low_return**2
  return hv_daily

hv_daily=parkinson_estimator(bitcoin_prices)
hv_daily
```

```python
plt.figure(figsize=(15,10))
plt.plot(hv_daily.index,hv_daily, label="Volatility paths",color="darkblue")
plt.legend()
plt.title("Parkinson estimator")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```

## Hidden Markov Models: Unveiling the Unseen

Imagine you're watching someone play a game where they flip a coin, but you can't see the coin itself. All you can see is whether they say "Heads" or "Tails." However, you notice something peculiar: sometimes they use a fair coin, and sometimes they switch to a biased coin that lands on heads more often. You don't know *when* they switch, only the sequence of "Heads" and "Tails" they announce.

This is the essence of a **Hidden Markov Model (HMM)**. It's a powerful statistical tool used to model systems where the underlying state (like which coin is being used) is *hidden* from you, but you can observe a sequence of events (the announced "Heads" or "Tails") that depend on those hidden states.

### What are the Key Components of an HMM?

An HMM has five main elements:

1.  **Hidden States (Q):** These are the unobservable states of the system. In our coin example, the hidden states would be "Fair Coin" and "Biased Coin."
2.  **Observation Symbols (V):** These are the things you *can* observe. In our example, they would be "Heads" and "Tails."
3.  **Transition Probabilities (A):** This matrix defines the probability of moving from one hidden state to another. For instance, what's the probability of switching from the "Fair Coin" state to the "Biased Coin" state? Or staying in the "Fair Coin" state?
        
4.  **Emission Probabilities (B):** This matrix defines the probability of observing a particular symbol given a specific hidden state. For example, if the hidden state is "Fair Coin," what's the probability of observing "Heads"? If it's "Biased Coin," what's the probability of "Heads"?
5.  **Initial State Probabilities (π):** This vector specifies the probability of starting in each hidden state. What's the probability that the game begins with the fair coin, or with the biased coin?

### How do HMMs Work?

HMMs solve three fundamental problems:

1.  **Evaluation (Likelihood Problem):** Given an HMM and a sequence of observations, what is the probability that this sequence was generated by this specific HMM? (e.g., How likely is it to see "Heads, Tails, Heads" if we know the coin transition and emission probabilities?)
2.  **Decoding (Most Likely State Sequence):** Given an HMM and a sequence of observations, what is the most likely sequence of *hidden states* that produced these observations? (e.g., What was the most likely sequence of coin switches given the "Heads, Tails, Heads" output?) The **Viterbi Algorithm** is commonly used to solve this problem.
3.  **Learning (Parameter Estimation):** Given an HMM and a sequence of observations, how can we adjust the HMM's parameters (transition and emission probabilities) to best explain the observed data? This is often solved using the **Baum-Welch Algorithm** (a variant of EM algorithm).

### Where are HMMs Used?

HMMs are incredibly versatile and find applications in numerous fields:

*   **Speech Recognition:** Predicting the sequence of words (hidden states) given the acoustic signal (observations).
*   **Bioinformatics:** Identifying genes in DNA sequences, protein secondary structure prediction.
*   **Natural Language Processing:** Part-of-speech tagging, named entity recognition.
*   **Finance:** Modeling stock market fluctuations.
*   **Weather Forecasting:** Predicting weather patterns based on atmospheric observations.
*   **Gesture Recognition:** Interpreting human gestures from sensor data.

### Advantages and Limitations

**Advantages:**

*   **Handles sequential data:** Excellent for modeling time-series data where the order of events matters.
*   **Probabilistic framework:** Provides a principled way to deal with uncertainty.
*   **Interpretable:** The hidden states and probabilities can often be mapped back to real-world concepts.

**Limitations:**

*   **Assumes Markov property:** The current hidden state only depends on the previous hidden state (memoryless). This might not always hold true in complex real-world systems.
*   **Fixed number of hidden states:** You need to pre-define the number of hidden states, which can be challenging.
*   **Computational complexity:** Training HMMs can be computationally intensive, especially with many states or long observation sequences.


#Mathematical explaination
Hidden Markov Models detect market regimes by modeling returns as being generated by a small number of unobserved states (e.g., low- vs high-volatility), learning how the system switches between them and then inferring the most likely current state and next-state probabilities from observed features like returns and ranges. In practice, an HMM is fit to historical **features**, the most likely *past state path* is decoded and *filtered state probabilities* are used to adapt exposure, position sizing or risk limits in a regime-aware way.

So, an HMM is a probabilistic state-space model where a *latent state sequence* follows a Markov chain and emits observable data according to state-specific distributions. For regime detection, the **hidden states** represent **market conditions** (e.g., bull/low-vol, bear/high-vol) that govern the distribution of observable returns-related features.

### Model components
The HMM is defined by initial state probabilities $ \pi $, a transition matrix $ A=[a_{ij}] $ for switching among states and emission distributions $ p(x_t \mid S_t=i;\theta_i) $ that model observed features given the state. Common choices set emissions as Gaussian with state-specific means and covariances and use inputs such as log returns and daily high–low range to capture level and volatility information.
>They’re called “emissions” because in an HMM each hidden state generates (emits) an observable output according to a state‑dependent probability distribution, so the observation is viewed as being emitted by the active state at that time. In the literature these are also called observation or output probabilities, emphasizing that they link hidden states to what is observed. Formula: $P(X_t=x|S_j=i)$

The Markov structure allows the model to capture persistence and switching between regimes via learned transition probabilities.

### Training and decoding
Parameters are typically learned by the Baum–Welch algorithm (an EM procedure using the forward–backward recursions) to maximize the likelihood of the observed feature sequence under the HMM.


> The goal of the algorithm is to maximize the data likelihood under an HMM by iteratively refining initial probabilities $ \pi $, transition matrix $A$ and emission parameters, using EM specialized to HMMs. It has two steps:
- E‑step: run forward–backward to compute posterior state probabilities $ \gamma_t(i)=P(S_t=i\mid x_{1:T}) $ and expected transitions $ \xi_t(i,j)=P(S_t=i,S_{t+1}=j\mid x_{1:T}) $ for sufficient statistics.
- M‑step: update $ \pi, A $ and emission parameters from these expectations to increase the sequence likelihood; iterate until the log‑likelihood converges.

Since it has two recursions, the E-step is calculated in this way:
- Forward (filtering) recursions:  
  $ \alpha_1(j)=\pi_j\,b_j(x_1),\ \ \alpha_t(j)=\big[\sum_i \alpha_{t-1}(i)a_{ij}\big]\,b_j(x_t) $, and $ p(x_{1:T})=\sum_j \alpha_T(j) $ as the sequence likelihood.
  $$
  P(S_t=j\mid x_{1:t})=\frac{\alpha_t(j)}{\sum_{i=1}^K \alpha_t(i)}\,.
  $$
- Backward recursions:
  It computes retrospective regime probabilities $P(S_t=j\mid x_{1:T})$ for clean labels and better parameter/diagnostic estimates.
  - Define backward messages $$ \beta_t(i)=p(x_{t+1:T}\mid S_t=i) $$ with initialization and recursion:
  $$
  \beta_T(i)=1,\qquad
  \beta_t(i)=\sum_{j=1}^K a_{ij}\,b_j(x_{t+1})\,\beta_{t+1}(j)\,.
  $$  to propagate future evidence.
- Posteriors and pairwise posteriors:  
  $ \gamma_t(i)=P(S_t=i\mid x_{1:T})=\dfrac{\alpha_t(i)\beta_t(i)}{\sum_j \alpha_t(j)\beta_t(j)} $ and $ \xi_t(i,j)=P(S_t=i,S_{t+1}=j\mid x_{1:T})=\dfrac{\alpha_t(i)a_{ij}b_j(x_{t+1})\beta_{t+1}(j)}{\sum_{p,q}\alpha_t(p)a_{pq}b_q(x_{t+1})\beta_{t+1}(q)} $.

Instead, the M step:
- Initial probabilities: $ \pi_i^{\text{new}}=\gamma_1(i) $ (or averaged over sequences).
- Transitions: $ a_{ij}^{\text{new}}=\dfrac{\sum_{t=1}^{T-1}\xi_t(i,j)}{\sum_{t=1}^{T-1}\gamma_t(i)} $ (expected transitions over expected time in state i).
- Gaussian emissions for regime detection features $x_t\in\mathbb{R}^d$:  
  $ \mu_i^{\text{new}}=\dfrac{\sum_t \gamma_t(i)\,x_t}{\sum_t \gamma_t(i)} $,  
  $ \Sigma_i^{\text{new}}=\dfrac{\sum_t \gamma_t(i)\,(x_t-\mu_i^{\text{new}})(x_t-\mu_i^{\text{new}})^\top}{\sum_t \gamma_t(i)} $


- Stop when the increase in log‑likelihood falls below a tolerance or after a max iteration count; EM guarantees non‑decreasing likelihood and typically converges to a local optimum in HMMs.
- Memory/time can be optimized with linear‑memory variants or incremental updates, but the core E/M steps remain the same conceptually.

Once trained, the Viterbi algorithm decodes the most likely historical state sequence, while filtered or smoothed state probabilities provide uncertainty-aware estimates of the current regime and the likelihood of transitioning to another regime. These decoded paths and probabilities form the basis for regime-aware signals, filters or risk overlays.

> The Viterbi returns the maximum a posteriori (MAP) sequence of hidden states given the observed returns features and the fitted model $ \hat{\pi}, \hat{A}, \hat{b}_i(\cdot) $. Unlike filtering or smoothing, which give probabilities per time, Viterbi gives one globally consistent path that maximizes the joint posterior over the whole sequence.
Therefore:
- Input: a trained HMM with states, initial probabilities $ \pi $, transition matrix $ A=[a_{ij}] $ and emission distributions $ b_i(x) $, plus an observation sequence $ x_{1:T} $ from returns‑based features for regime detection.
- Output: the most likely hidden regime sequence $ \hat{s}_{1:T} $ (the Viterbi path) explaining $ x_{1:T} $ under the model, useful for clean regime labels and interpretation of episodes.

- Define the Viterbi score as the best path probability ending in state $ j $ at time $ t $:  
  $$
  v_t(j)=\max_{s_{1:t-1}} P(s_{1:t-1},x_{1:t},S_t=j\mid \lambda)\,,
  $$
  with initialization $$ v_1(j)=\pi_j\,b_j(x_1) $$ and recursion  
  $$
  v_t(j)=\max_{i} \big[v_{t-1}(i)\,a_{ij}\,b_j(x_t)\big] \,,
  $$
  and termination by selecting $ \hat{S}_T=\arg\max_j v_T(j) $.
- Store backpointers  
  $$
  \psi_t(j)=\arg\max_{i} \big[v_{t-1}(i)\,a_{ij}\,b_j(x_t)\big]
  $$
  at each cell, then backtrack $$ \hat{S}_{t}=\psi_{t+1}(\hat{S}_{t+1}) $$ to recover the full path $$ \hat{s}_{1:T} $$ efficiently.

### Log‑space version
- To avoid underflow, work with log‑scores $ \ell_t(j)=\log v_t(j) $ so the recursion becomes  
  $
  \ell_t(j)=\max_i \big[\ell_{t-1}(i)+\log a_{ij}+\log b_j(x_t)\big],\ \ \ell_1(j)=\log\pi_j+\log b_j(x_1)\,,
  $
  with the same backpointers and backtracking procedure for the MAP path.

### Relation to forward/backward
- Forward–backward sums over all paths to compute per‑time posteriors and the likelihood, whereas Viterbi maximizes over paths to find a single best sequence; algorithmically the trellis is identical except that sums are replaced by maxima and backpointers are recorded in Viterbi.
- Practically, use Viterbi for crisp regime labels and interpretation and use filtered or smoothed probabilities for uncertainty‑aware execution, diagnostics and EM re‑estimation.



### Initialization
- Select regimes $K$ (start with 2–3) and emission family (Gaussian is common; consider heavier‑tailed robustness via features).  
- Initialize $ \pi $ (uniform), $ A $ (near‑diagonal for persistence) and emission moments (k‑means or sample stats per tentative cluster).

### Train (Baum–Welch)
- Run EM until the increase in log‑likelihood is below tolerance; use scaled forward–backward or log‑space to avoid underflow.  
- Save fitted parameters $ \hat{\pi}, \hat{A}, \hat{\theta} $ and the training log‑likelihood evolution for diagnostics.

### Decode and infer
- Viterbi decode: compute the MAP regime path $ \hat{s}_{1:T} $ for clear, single‑label timelines and qualitative interpretation.  
- Filtering: compute $P(S_t=i\mid x_{1:t})$ for live, causal regime probabilities; smoothing: compute $P(S_t=i\mid x_{1:T})$ for clean historical labels and better parameter analysis.

### Regime interpretation
- Summarize each state’s conditional mean, volatility, and other feature moments to label regimes (e.g., low‑vol/positive drift vs high‑vol/negative drift).  
- Align decoded regimes with known episodes (crashes, rallies) to validate economic plausibility and narrative consistency.

### Forecasting and control
- One‑step regime forecast: propagate filtered probabilities with $ p_{t+1\mid t}=p_t A $ to anticipate transitions and adjust risk preemptively.  
- Compute expected next‑day dispersion or drift by mixing state‑conditional forecasts with $ p_{t+1\mid t} $.


### Validation and selection
- Use likelihood, AIC/BIC, and out‑of‑sample performance to choose $K$, features and emission settings; prefer parsimonious models for stability.  
- Backtest strategies with regime filters; compare to baselines without HMM gating; examine drawdowns, turnover and robustness across windows.

### Deployment
- Switch to filtered probabilities for live decisions; keep Viterbi and smoothed probabilities for reporting and offline diagnostics.  
- Choose initial probabilities for live filtering: uniform, EM‑learned $ \gamma_1 $ or the stationary distribution solving $ \pi^\top=\pi^\top A $ for stable restarts.

  - Uniform: set $ \pi_i = 1/K $; simple and neutral, good when starting from scratch or when EM will immediately adapt during training.  
  - EM‑learned: after one E‑step on the training set, set $ \pi_i = \gamma_1(i)=P(S_1=i\mid x_{1:T}) $; this matches the training sample’s start and is natural when carrying the model directly into validation.  
  - Stationary: compute $ \pi $ from $A$ and use it for stable restarts and live filtering when the calendar start is arbitrary; it avoids biased early‑period inferences caused by an unlucky initialization.

> A stationary distribution is a probability vector $ \pi $ over regimes that does not change after applying the transition matrix $A$, i.e., it solves $ \pi^\top = \pi^\top A $; it captures the long‑run fraction of time spent in each regime and is a stable, data‑driven choice for initializing live filtering when the start time is arbitrary. Uniform and EM‑learned $\gamma_1$ are also valid: uniform is a neutral default and $\gamma_1$ matches the training sample’s start; all three are reasonable depending on the use case. For an ergodic Markov chain (irreducible and aperiodic), the stationary distribution exists and is unique, and it equals the long‑run fraction of time the chain spends in each state.
> We compute it by using:
- Linear system: solve $ \pi^\top = \pi^\top A $ plus $ \sum_i \pi_i = 1 $; equivalently solve $ (A^\top - I)\,\pi = 0 $ with the normalization constraint.  
- Eigenvector view: $ \pi^\top $ is the left eigenvector of $A$ associated with eigenvalue $1$; compute that eigenvector and normalize it to sum to $1$.  
- Power iteration: start from any distribution $p^{(0)}$, iterate  $p^{(n+1)} = p^{(n)} A $ and renormalize; for ergodic chains, $ p^{(n)} \to \pi $.  
- Two‑state closed form: for $ A=\begin{bmatrix}1-p & p\\ q & 1-q\end{bmatrix} $ with $p,q>0$, the stationary distribution is $ \pi=\big[ \tfrac{q}{p+q},\ \tfrac{p}{p+q} \big] $.



### Advanced options
- Time‑varying transitions (TVTP): model $A_t$ with covariates (e.g., macro, funding, on‑chain); use filtering for live $P(S_t)$ and maintain a covariate‑aware initialization.  
- Markov‑switching AR or state‑dependent regressions if serial correlation is material; mixture emissions if single Gaussians are too brittle.  
- Bayesian HMMs (Dirichlet priors on $ \pi, A $) for full uncertainty quantification when risk decisions require posterior intervals.

#Step 4:Modelling purpose: GARCH

The GJR-GARCH model is a volatility model that extends GARCH by adding an asymmetry term so that negative shocks raise future variance more than equal‑sized positive shocks, capturing the leverage effect observed in many financial return series. It is widely used when volatility clusters over time and bad news tends to have a disproportionate impact on risk estimates relative to good news.

Classical GARCH models capture volatility clustering but assume that up and down moves of the same magnitude have identical effects on future volatility, which often understates the empirical “bad news increases volatility more” pattern in equity and crypto returns. GJR-GARCH fixes this by activating an extra variance term only after negative shocks, making volatility more sensitive to losses than to gains of the same size.

### Model setup
---
A common observation equation sets returns as $ r_t = \mu_t + \epsilon_t $ with innovations $ \epsilon_t = \sigma_t z_t $, where $ z_t $ is i.i.d. with mean 0 and variance 1 and $ \sigma_t^2 $ is the conditional variance process to be modeled. The conditional variance in GJR‑GARCH(1,1) is specified as $ \sigma_t^2 = \omega + \alpha\,\epsilon_{t-1}^2 + \gamma\,\epsilon_{t-1}^2 \mathbb{I}\{\epsilon_{t-1}<0\} + \beta\,\sigma_{t-1}^2 $.

Here $ \mathbb{I}\{\cdot\} $ is an indicator equal to 1 if the previous shock was negative and 0 otherwise, so the variance reacts as $ \alpha $ to positive shocks and $ \alpha+\gamma $ to negative shocks of the same magnitude.

### Asymmetry and leverage
If $ \gamma>0 $, a negative return shock increases next‑period variance more than a positive shock with the same absolute size, which is consistent with leverage and volatility feedback effects documented in equities and other risk assets. This asymmetry is why GJR‑GARCH is also called a threshold GARCH (TGARCH), reflecting that the variance update crosses a threshold when the sign of the shock is negative.

### Parameter constraints
Typical constraints ensure positivity and well‑behaved dynamics: $ \omega>0 $, $ \alpha\ge 0 $, $ \beta\ge 0 $, and $ \gamma\ge 0 $, which guarantee that conditional variance stays non‑negative. Covariance stationarity requires that the combined persistence remains below one, with conditions expressed in terms of $ \alpha $, $ \beta $, and the asymmetry load $ \gamma $ under assumptions about the innovation distribution and indicator definition.

Parameters are commonly estimated by maximum likelihood, choosing a residual distribution such as Gaussian or Student‑t, with the latter often fitting heavy‑tailed financial returns better in practice. In software, GJR‑GARCH(1,1) corresponds to one ARCH lag, one asymmetry (threshold) lag, and one GARCH lag, and it is standard in academic and practitioner toolkits for risk and volatility forecasting.

### Forecasting behavior
One‑step‑ahead forecasts update $ \sigma_{t+1}^2 $ from the latest shock and variance, while multi‑step forecasts iterate the recursion and typically revert toward a long‑run variance level implied by the parameters when persistence is less than one. Because negative shocks load more heavily through $ \gamma $, stress episodes can produce sharper short‑run volatility spikes than symmetric GARCH would imply, which improves risk measurement during downturns.

GJR‑GARCH is preferred over symmetric GARCH when data exhibit volatility clustering and stronger variance responses to negative returns. If diagnostics indicate significant asymmetry in standardized residuals or news‑impact asymmetry, GJR‑GARCH typically outperforms linear GARCH in out‑of‑sample volatility forecasting and risk estimation.

##Tests for asymmetry of the time series


---

Instead of directly use GARCH or even asymmetric GARCH, it is necessary to perform tests for extracting asymmetry in time series.


### 1. **Volatility Asymmetry Tests**

* **Engle–Ng (1993) tests for asymmetry in volatility**:

  * **Sign bias test** – checks whether positive/negative shocks have different effects on volatility.
  * **Negative size bias test** – tests whether the magnitude of negative returns affects volatility differently.
  * **Positive size bias test** – same as above, for positive returns.
* These are often applied in the context of GARCH-type models (EGARCH, GJR-GARCH).

Suppose you have returns $r_t$, and a fitted volatility model with conditional variance $\hat{\sigma}_t^2$. Define the standardized residuals:

$$
\epsilon_t = \frac{r_t}{\hat{\sigma}_t}
$$

Then construct dummy variables:

* $d_t^- = 1$ if $\epsilon_{t-1} < 0$, else 0
* $d_t^+ = 1$ if $\epsilon_{t-1} > 0$, else 0

Tests:

* **Sign Bias Test**: regress squared residuals on the dummy for negative shocks:

  $$
  \epsilon_t^2 = \alpha + \beta d_{t-1}^- + u_t
  $$

  Null: $\beta = 0$ (no sign effect). If rejected → negative vs. positive shocks matter differently.

* **Negative Size Bias Test**: regress on $d_{t-1}^- \cdot \epsilon_{t-1}$.

* **Positive Size Bias Test**: regress on $d_{t-1}^+ \cdot \epsilon_{t-1}$.

* **Joint Test**: regress on all three dummies together.

If coefficients are significant → evidence of asymmetric volatility.

---

## 2. **Nonlinear & Threshold Models**

* **Threshold Autoregression (TAR)**:

  $$
  y_t =
  \begin{cases}
  \phi_1 y_{t-1} + \varepsilon_t & \text{if } y_{t-d} \leq \tau \\
  \phi_2 y_{t-1} + \varepsilon_t & \text{if } y_{t-d} > \tau
  \end{cases}
  $$

  * $\tau$ = threshold (often 0 or mean).
  * $d$ = delay parameter.
    Test: check if $\phi_1 \neq \phi_2$. If yes → the dynamics depend on being in “low” vs. “high” regime.

* **STAR model (Smooth Transition AR)**: smoother version, transition governed by logistic function. Tests asymmetry in how dynamics change gradually with the variable’s level.

* **Threshold Unit Root Tests (Caner–Hansen, 2001)**: test whether a time series has a unit root in one regime but is stationary in another. Useful for asymmetric persistence.

---

##  3. **Distributional Asymmetry**

* **Skewness test**: skewness statistic

  $$
  \text{Skewness} = \frac{E[(X-\mu)^3]}{\sigma^3}
  $$

  * Positive skew → heavy right tail.
  * Negative skew → heavy left tail.

* **Jarque–Bera test**:

  $$
  JB = \frac{n}{6}\left(S^2 + \frac{(K-3)^2}{4}\right)
  $$

  where $S$ = skewness, $K$ = kurtosis.

  * Tests jointly for non-normal skew/kurtosis.

* **Rolling skewness**: compute skewness on a moving window to see if asymmetry changes over time.

---

## 4. **Leverage Effect Tests**

The **leverage effect** is when negative returns increase future volatility more than positive returns of the same size.

Ways to test:

* Compute correlation between past returns and future squared returns (proxy for volatility):

  $$
  \rho = \text{Corr}(r_{t-1}, r_t^2)
  $$

  If negative → evidence of leverage.

* In GARCH models, check if asymmetric terms (e.g., in EGARCH or GJR-GARCH) are significant.

---

## 5. **Nonparametric & Spectral Tests**

* **Quantile Autoregression (Koenker–Xiao)**:
  Estimate autoregression not just for the mean, but for different quantiles. Example:

  $$
  Q_{y_t}(\tau | y_{t-1}) = \alpha(\tau) + \beta(\tau) y_{t-1}
  $$

  If $\beta(\tau)$ changes across quantiles → asymmetry in dynamics.

* **Bispectrum analysis**:
  The bispectrum is a third-order spectral measure that detects nonlinearity and asymmetry.

  * Symmetric linear processes → bispectrum is zero.
  * Nonzero bispectrum → evidence of asymmetric dependence.

---

## 6. **Directional Dependence**

* **Granger causality in quantiles**:
  Test whether “bad news” (negative returns) Granger-cause volatility differently than “good news.”

* **Asymmetric cross-correlation**:

  * Split returns into positive and negative parts:

    $$
    r_t^+ = \max(r_t, 0), \quad r_t^- = \min(r_t, 0)
    $$

  * Compute correlations with future volatility separately:

    * $Corr(r_t^+, \sigma_{t+1}^2)$
    * $Corr(r_t^-, \sigma_{t+1}^2)$

  * If these differ → asymmetric response of volatility.

#Model construction
We define a very basic function to apply a GARCH model to each regime.

The function is `parameters_garch()` and it takes as argument the time series, the p, o, and q parameters and dist that allows us to either choose a simple normal distribution or choose t-student for the asymmetric behaviour.

This function returns the GARCH model with the best parameters according the AIC and BIC criterion.

```python
def gjr_garch_ll(params, returns):
    mu, omega, alpha, gamma, beta = params
    T = len(returns)

    eps = returns - mu
    sigma2 = np.zeros(T)

    sigma2[0] = np.var(returns)

    for t in range(1, T):
        indicator = 1 if eps[t-1] < 0 else 0
        sigma2[t] = omega + alpha*eps[t-1]**2 + gamma*indicator*eps[t-1]**2 + beta*sigma2[t-1]

    loglik = -0.5 * np.sum(np.log(2*np.pi) + np.log(sigma2) + (eps**2)/sigma2)
    return loglik


def parameters_garch(returns, p_values, o_values, q_values, dist):
    '''This function is useful for finding the best combination of parameters
    for the model according the AIC and BIC criterions.
    returns: time series;
    p_values, o_values, q_values: list of possible values of the params;
    dist: string ("normal" or "t")'''
    best_model = None
    best_res = None
    best_aic = np.inf
    best_spec = None
    best_bic= np.inf

    for p, o, q in itertools.product(p_values, o_values, q_values):
        try:
            model = arch_model(
                returns,
                vol='GARCH', #vol modelled using GARCH, GJR-GARCH with t
                p=p, o=o, q=q,
                dist=dist,
                mean='AR' #modelled using the AR
            )
            res = model.fit(disp="off")
            aic = res.aic
            bic = res.bic
            print(f"Done with model: p={p}, o={o}, q={q}, AIC={aic}")

            if aic < best_aic:
                best_aic = aic
            if bic < best_bic:
                best_bic = bic
                best_model = model      # save the right model
                best_res = res          # save the results
                best_spec = (p, o, q)   # save the specification
        except Exception as e:
            print(f"Error with parameters p={p}, o={o}, q={q}: {e}")
            continue

    return best_aic, best_model, best_res, best_spec


p_values=[1,2]
o_values=[0,1]
q_values=[1,2]

best_aic_0, right_model_0, best_model_0, best_params = parameters_garch(returns_regime_0["Close"],p_values,o_values,q_values, 't')
print(f"\nBest AIC: {best_aic_0}")
print("Best Model Summary:")
print(best_model_0.summary())
```

```python
horizon_0=30
horizon_1=30
def forecasting_values(data,horizon,start,regime):
  '''this function plots the forecasted value for the time series according
  the identified regime.
  data: time series;
  horizon: number of forecasted paths;
  start: initial date;
  regime: string'''

  #initial forecasts according to the model and horizon
  forecasts=data.forecast(horizon=horizon,start=start, method="bootstrap")
  forecasts_index=forecasts.mean.index
  forecasts_mu=forecasts.mean.iloc[-1].values
  forecasts_var= forecasts.variance.iloc[-1].values
  sqrt=np.sqrt(forecasts_var)
  z=1.96
  #simulate the returns using the forecasted mu and sqrt
  returns= forecasts_mu+sqrt*np.random.randn(horizon)
  #identify the confidence intervals
  upper= returns+z*sqrt
  lower= returns-z*sqrt

  #plot it
  plt.figure(figsize=(10,6))
  plt.plot(range(1,horizon+1), returns, color="mediumblue",marker='o', label=f'Simulated returns for {regime}')
  plt.fill_between(range(1,horizon+1), lower, color='skyblue', alpha=0.3, label='95% CI')
  plt.fill_between(range(1,horizon+1), upper, color='darkcyan', alpha=0.3, label='95% CI')
  plt.plot(range(1,horizon+1), forecasts_mu, color='red', linestyle='--', alpha=0.3, label='Forecasted mean')
  plt.xlabel('Steps ahead')
  plt.ylabel('Returns')
  plt.title(f'GARCH Forecast: Simulated Returns with 95% Confidence for {regime}')
  plt.legend()
  plt.show()

  return returns,upper,lower,forecasts,forecasts_index

returns_0,upper_0,lower_0,forecasts_0,index_0=forecasting_values(best_model_0,horizon_0,"2025-08-01","Regime 0")
returns_1,upper_1,lower_1,forecasts_1,index_1=forecasting_values(best_model_1,horizon_1,"2025-08-01","Regime 1")



```

Grange causality means that past values of $x_2$ have a

values of $x_1$ into account as regressors.
If p-value is lower than 0.05, we reject the null hypothesis

The null hypothesis for all four test is that the coefficients
corresponding to past values of the second time series are zero.

- in the second lag is the same of the lag 1;
