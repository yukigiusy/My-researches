#Black-Scholes model

Black-Scholes model is a financial derivative model, used to price the call and put prices, in the specific used in the *European options*.

##Assumptions
There are different assumptions to exploit before writing the formula.
1. **Lognormal**: the stock price follows a lognormal distribution. It basically means that there is high risk on the downturn movement;
2. **frictionless market**: there are no transaction costs and the assets are divisible;
3. **costant** risk free interest rates;
4. **no dividends** (but it exists the Merton model that includes them;
5. **volatility** is assumed to be constant;
6. **no arbitrage opportunities**.



## Log-normal stock price
The **lognormal distribution** is a probability distribution of a **random variable whose logarithm is normally distributed**. If a random variable $X$ follows a lognormal distribution, then:

$
Y = \ln(X) \sim \mathcal{N}(\mu, \sigma^2)
$
where $Y$ follows a normal distribution with mean $ \mu $ and variance $ \sigma^2$.

### **Mean and Variance of the Lognormal Distribution**
If $X \sim \text{Lognormal}(\mu, \sigma^2)$, then:

1. **Mean (Expected Value):**
   $
   \mathbb{E}[X] = e^{\mu + \frac{\sigma^2}{2}}
   $

2. **Variance:**
   $
   \text{Var}(X) = e^{2\mu + \sigma^2} (e^{\sigma^2} - 1)
   $
These expressions show that the lognormal distribution is **skewed** and strictly **positive**, which makes it useful for modeling asset prices, income distributions, and other financial and natural phenomena.

In the specific it is used in order to model the stock price.

## Brownian motion and geometrical Brownian motion


---


In continuous time we model the change in the price of our financial assets by establishing that:
- $S_t$ is the future price and the current time is t=0;
- $S_t$ is defined when t > 0;
- there is no uncertainty.

The, we define that the growth rate is the proportional change per unit time and for this reason we write:
$\frac{d S_t}{dt}= \mu S_t$ and after the calculation of the ordinary differential equation, we end up with $S_t=S_0 e^\mu$.

But since there is not only the deterministic component, we need to add also the stochastic component and it is modelled by using a stochastic differential equation.

Now we have $\frac{d S_t}{S_t}= \mu d_t + \sigma dW_t$, where $W_t$ is the Wiener process.

The Brownian motion is a continuous process with stationary, independent normally distributed increments (the total change of the process in an infinitesimal period). It is used to model the stochastic part.

It starts at zero.

### Assumptions:


---


1. **the increments are independent**;
2. they are **normally distributed** and it holds for arbitrary time intervals t and s;
3. it has continuous and **not differentiable** paths. This is connected to the fact that we think at the derivatives like something that linearize our function and we calculate the rate of change; the BM, being an infinite summation of random walks, could not be linearized;
4. Brownian motion is a martingale, so $E(X_t | ℱ_s)= W_s $ for s < t
5. the quadratic variation, that has important implication in the calculation of the Ito's lemma.

The **quadratic variation** of a process $ X_t$ over the interval $[0,T]$ is defined as:

$
[X]_T = \lim_{|\Pi| \to 0} \sum_{i=1}^{n} (X_{t_i} - X_{t_{i-1}})^2
$

where $\Pi = \{0 = t_0 < t_1 < \dots < t_n = T\} $ is a partition of $ [0,T] $ and $ |\Pi| $ is the mesh size (the largest subinterval length).

For **standard Brownian motion** \( B_t \), we show that its quadratic variation is:$[B]_T = T$

**Brownian motion has quadratic variation $[B]_T = T$**, meaning its squared increments, that are independent and for that reason we can sum the different variances of the increments, accumulate linearly over time.


5. covariance property: $\mathbb{E}[B_s B_t] = \min(s, t)$
$
B_t = B_s + (B_t - B_s).
$

Since Brownian motion has independent increments, the term $ B_t - B_s $ is independent of $B_s $ and is normally distributed with:

$
B_t - B_s \sim \mathcal{N}(0, t - s).
$

Using the linear decomposition:

$
\mathbb{E}[B_s B_t] = \mathbb{E}[B_s (B_s + (B_t - B_s))].
$
Expanding the expectation:

$
\mathbb{E}[B_s B_t] = \mathbb{E}[B_s^2] + \mathbb{E}[B_s (B_t - B_s)].
$

We now evaluate each term:

1. **First term: $ \mathbb{E}[B_s^2] $**  
   Since $B_s \sim \mathcal{N}(0, s)$, we use the variance formula:

   $
   \mathbb{E}[B_s^2] = s.
   $

2. **Second term: $\mathbb{E}[B_s (B_t - B_s)] $**  
   Since $B_s$ and $ B_t - B_s $ are independent, their expectation factorizes:

   $
   \mathbb{E}[B_s (B_t - B_s)] = \mathbb{E}[B_s] \cdot \mathbb{E}[B_t - B_s].
   $

   But we know that $B_s \sim \mathcal{N}(0, s)$ and $ B_t - B_s \sim \mathcal{N}(0, t-s) $, so their expectations are both 0:
   
   We are left with:
$
\mathbb{E}[B_s B_t] = s.
$

Since this holds for any $s \leq t $, and covariance is symmetric (i.e., $\mathbb{E}[B_s B_t] = \mathbb{E}[B_t B_s] $)* we conclude:

$
\mathbb{E}[B_s B_t] = \min(s,t).
$

- The function $\min(s,t)$ ensures that Brownian motion has a **stationary** structure, which is crucial in stochastic calculus and financial modeling.



---
*The covariance has the following properties:
- symmetry;
- positive definiteness;
- even property.


6. self-similarity: if you take a small segment of a Brownian path and scale it up appropriately, it statistically looks similar to a larger segment. This property makes it behave like a fractal, meaning it has structure at every scale, preventing smoothness.
Sure! Let's go deeper into the last part about the fractal-like nature of Brownian motion and Hölder continuity.  
The process exhibits self-similarity with exponent $H=1/2$

Hölder continuity provides a way to measure the smoothness of a function. A function $f(t)$ satisfies Hölder continuity of order $\alpha$ if there exists a constant $C$ such that for small $\Delta t $:
$
|f(t+\Delta t) - f(t)| \leq C (\Delta t)^\alpha.
$

- If a function is **differentiable**, then it must satisfy Hölder continuity with **$\alpha = 1 $**.  
- Brownian motion, however, satisfies Hölder continuity only for **$\alpha < 1/2 $**. This means that, as $\Delta t \to 0$, the increments shrink too slowly to allow for a well-defined derivative.


### **Itô's Lemma**  

Itô's Lemma is the stochastic calculus analog of the chain rule in classical calculus. It describes how a function of a **stochastic process** (such as Brownian motion) evolves over time.
Let $ B_t $ be a standard Brownian motion and let $ X_t $ be an Itô process:

$dX_t = \mu_t dt + \sigma_t dB_t$

where $\mu_t$ and $\sigma_t$ are adapted stochastic processes. If we define a function $f(t, X_t)$ that is **twice continuously differentiable** in $ x $ and continuously differentiable in $ t $, then $ f(t, X_t) $ follows the **stochastic differential equation (SDE)**:

$
df(t, X_t) = \left( \frac{\partial f}{\partial t} + \mu_t \frac{\partial f}{\partial x} + \frac{1}{2} \sigma_t^2 \frac{\partial^2 f}{\partial x^2} \right) dt + \sigma_t \frac{\partial f}{\partial x} dB_t.
$

This is the key result that allows us to compute the dynamics of a function of a stochastic process.


---


###Proof

Consider a function $f(t, X_t)$, where $X_t$ follows:

$
dX_t = \mu_t dt + \sigma_t dB_t.
$

Using a **Taylor expansion**,

$
df = \frac{\partial f}{\partial t} dt + \frac{\partial f}{\partial x} dX_t + \frac{1}{2} \frac{\partial^2 f}{\partial x^2} (dX_t)^2 + \mathcal{O}(dt^{3/2}).
$

Since $dX_t = \mu_t dt + \sigma_t dB_t $, we substitute:

$
df = \frac{\partial f}{\partial t} dt + \frac{\partial f}{\partial x} (\mu_t dt + \sigma_t dB_t) + \frac{1}{2} \frac{\partial^2 f}{\partial x^2} (\mu_t dt + \sigma_t dB_t)^2.
$

Now, using properties of **stochastic calculus**:
- $(dB_t)^2 = dt$,
- $dt^2 \approx 0$,  
- $dt dB_t \approx 0$,

we obtain:

$
df = \left( \frac{\partial f}{\partial t} + \mu_t \frac{\partial f}{\partial x} + \frac{1}{2} \sigma_t^2 \frac{\partial^2 f}{\partial x^2} \right) dt + \sigma_t \frac{\partial f}{\partial x} dB_t.
$


## Putting things together


---


The stock price is modelled by using the **geometric Brownian motion** (since it gives positive values and it respects the stock price structure):

$
S(t) = S_0 e^{X(t)}
$
where:  
- $X(t) = \sigma W(t) + \mu t $ is a **Brownian motion with drift**,  
- $S(0) = S_0 > 0 $ is the initial stock price.  
The GBM is a Markov process: the future given the present state is
independent of the past.

Taking the natural logarithm: $
X(t) = \ln\left(\frac{S(t)}{S_0}\right) = \ln S(t) - \ln S_0
$

which shows that: $
\ln S(t) = \ln S_0 + X(t)
$

Since $X(t)$ is normally distributed with mean $\mu t$ and variance $\sigma^2 t$, it follows that $S(t)$ has a **lognormal distribution** for each $t$.  

This formula is retrived by using the following stochastic differential equation (SDE):

$
dS_t = \mu S_t dt + \sigma S_t dW_t
$

where:  
- $ S_t $ is the stock price at time $ t $,
- $ \mu $ is the **drift rate** (expected return, it is the deterministic component and it tries to model the trend of the stock price),
- $\sigma$ is the **volatility** (standard deviation of returns),
- $ W_t$ is a **Wiener process (Brownian motion)**.;
- $\sigma S_t dW_t$ in the specific this part is called *diffusion term* and it captures the randomness part of the stock price.

Using **Itô’s Lemma**, we solve this equation by applying the change of variable:

$
d(\ln S_t) = \left(\mu - \frac{1}{2} \sigma^2 \right) dt + \sigma dW_t
$

Integrating from time $0 $ to $ t $:

$
\ln S_t = \ln S_0 + \left(\mu - \frac{1}{2} \sigma^2\right) t + \sigma W_t
$

Taking the exponential:

$
S_t = S_0 e^ \left( \left(\mu - \frac{1}{2} \sigma^2 \right) t + \sigma W_t \right)
$

We could think that the $d(\ln S_t)$ is $\Delta \ln P_t$ in the discrete equivalent as $\Delta \ln P_t = r + \epsilon_t$ that is a random walk with $\epsilon_t \sim \mathcal{N}(0, \sigma^2)$.

---


###Deterministic part

$
r = \mu + \frac{\sigma^2}{2}
$

then the expected stock price is given by:

$
\mathbb{E}[S(t)] = e^{rt} S_0
$

This means that, on average, the stock price grows like a **fixed-income security** with a continuously compounded interest rate $ r $.  
Obviously, the return of the fixed-income security is lower than the one of the stock since it is affected by the variability of the BM.

##Risk neutral measure
## **Radon-Nikodym Theorem and Change of Measure**
The **Radon-Nikodym theorem** is fundamental because it allows us to define a new probability measure $ Q $ (the **risk-neutral measure**) relative to the original probability measure $ P $ (the **real-world measure**). It guarantees the existence of a **Radon-Nikodym derivative**, which represents the **change of measure**.

### **Why is this important for Black-Scholes?**
- In financial markets, the **real-world measure $P$** describes asset price dynamics with an expected return $\mu$, meaning that asset prices incorporate a risk premium.
- The **risk-neutral measure $ Q $** is the measure under which all discounted asset prices are martingales.
- The Radon-Nikodym theorem ensures that we can switch between $ P $ and $ Q $ using a **density process**, often called the **state-price density** or **stochastic discount factor**:

  $
  \frac{dQ}{dP} = Z_T
  $

  where $ Z_T$ is the **Radon-Nikodym derivative (likelihood ratio)**.

- This change of measure allows us to **price options using risk-neutral expectations** rather than needing to model investors' risk preferences.

---

## **Girsanov’s Theorem and the Risk-Neutral Measure**
The **Girsanov Theorem** is a powerful result that describes how Brownian motion changes under a new probability measure. It tells us that **if we change from the real-world measure $P$ to the risk-neutral measure $ Q $, the drift of the process changes, but the volatility remains the same**.
**Novikov’s Condition** provides a sufficient condition to ensure that the change of measure is valid.

Let $( \Omega, \mathcal{F}, P ) $ be a probability space with a standard Brownian motion $ B_t^P $ under the measure $P$. Suppose we define a new process:

$
\theta_t = (\theta_s)_{0 \leq s \leq t}
$

such that the **exponential martingale** (also called the **Radon-Nikodym derivative**) is given by:

$
L_t = \exp \left( -\int_0^t \theta_s dB_s^P - \frac{1}{2} \int_0^t \theta_s^2 ds \right).
$

Then, if $(L_t) $ is a martingale, we can define a new probability measure $ Q $ by:
$
\frac{dQ}{dP} = L_T.
$
Therefore, under $ Q$, the process $B_t^Q$ defined as:
$
B_t^Q = B_t^P + \int_0^t \theta_s ds
$ is a **Brownian motion under $Q$**.

For the change of measure to be valid (i.e., for $ L_t$ to be a true martingale), we need to ensure that the expectation of $ L_t $ is 1. A **sufficient condition** for this is **Novikov’s Condition**:
$
\mathbb{E}^P \left[ \exp \left( \frac{1}{2} \int_0^T \theta_s^2 ds \right) \right] < \infty.
$

### **Why is Novikov’s Condition Important?**
- It guarantees that $(L_t)$ is a **martingale**, which is necessary to define $Q $.
- Without this condition, $L_t $ might explode (i.e., have infinite expectation), making $ Q $ an **improper** measure.
---

### **Girsanov’s Theorem in the BSM Model**
Consider the **real-world price dynamics** of a stock $ S_t $:

$
dS_t = \mu S_t dt + \sigma S_t dB_t^P,
$

where:
- $\mu$ is the real drift (expected return).
- $\sigma$ is the volatility.
- $B_t^P$ is a Brownian motion under $ P $.

To price options, we move to the **risk-neutral measure $ Q $**, where the discounted stock price follows a martingale. Girsanov’s theorem tells us that under $ Q$, the new Brownian motion $B_t^Q$ satisfies:

$
dB_t^Q = dB_t^P - \theta dt,
$

where $\theta = \frac{\mu - r}{\sigma} $ is the **market price of risk** or the Sharpe ratio.

Thus, under $ Q $, the stock follows:
$
dS_t = r S_t dt + \sigma S_t dB_t^Q.
$

- The drift **changes from $ \mu $ to $ r$**, the risk-free rate.
- The volatility $\sigma $ remains unchanged.
- This transformation **eliminates the risk premium** and allows us to **price options using discounted expectations under $Q$**.

---

## **Final Connection to Black-Scholes**
Now, using Girsanov's theorem:
- We express option prices as risk-neutral expectations.
- The famous Black-Scholes formula comes from solving the PDE under the risk-neutral measure.

$
C(S_t, t) = e^{-r(T-t)} \mathbb{E}^Q [\max(S_T - K, 0) | \mathcal{F}_t].
$

This means that option prices can be computed by taking the expectation under the **risk-neutral measure** $Q$, where the drift of $S_t$ is $ r $ instead of $\mu $.




### Dividends: Merton modification


---


The risk-neutral stochastic differential equation (SDE) for the stock price under dividends is:

$
dS_t = (\mu - q) S_t dt + \sigma S_t dW_t
$

where:  
- $ S_t $ is the stock price at time $t$.  
- $\mu$ is the drift (expected return).  
- $q$ is the continuous dividend yield.  
- $\sigma$ is the volatility.  
- $dW_t$ is a Wiener process.  

We subtract the q because once the dividends are paid, the stock price drops since a dividend is essentially a transfer of value from the company to its shareholders and now the company holds less assets.

##BSM Formula
The Black-Scholes formula for a European call option price is given by:

$
C = N(d_1) S_t - N(d_2) K e^{-rt}
$

where:

$
d_1 = \frac{\ln \frac{S_t}{K} + \left( r + \frac{\sigma^2}{2} \right) t}{\sigma \sqrt{t}}
$

$
d_2 = d_1 - \sigma \sqrt{t}
$

Here:
- $ C $ = call option price  
- $S_t $ = current stock price  
- $K$ = strike price  
- $r$ = risk-free interest rate  
- $t$ = time to expiration (in years)  
- $\sigma$ = volatility of the stock  
- $N(d)$ = cumulative distribution function (CDF) of the standard normal distribution

###Elements
- S: is the stock price. If it increases, the call option becomes more ITM and it increases the option value (the opposite for the put option);

- K: is the strike price. We need a low K for ITM call options if we want a high price for the option;

- r: is the risk-free interest rates. It is applied on "K", so if it is high, it will exponentially reduce K and it increases the option value;

- T: is the time to maturity. The longer is, the higher is the option value;

- $\sigma$: is the implied volatility. The higher is, the higher is the option value.

```python
expiration= NVDA.options
chosen_date= expiration
expiration[3]
```

```python
options= NVDA.option_chain(expiration[3])
call_op= options.calls.dropna()
call_op
```

```python
def time_to_maturity(expiration, trade_day):
  exp = pd.to_datetime(expiration).tz_localize(None) # Make exp timezone-naive
  trade = pd.to_datetime(trade_day).tz_localize(None) # Make trade timezone-naive
  time_diff = (exp - trade).days
  time_diff = time_diff / 365
  return time_diff

first_time = time_to_maturity(expiration[3], call_op['lastTradeDate'].iloc[0])
first_time
```

```python
def BS_call(S,K,r,sigma,T):
  d1 = (np.log(S/K) + (r + (sigma**2)/2) * T) / (sigma * np.sqrt(T))
  d2= d1-sigma*np.sqrt(T)
  return S*N(d1)-K*np.exp(-r*T)*N(d2)
```

```python
def impl_vol_cal(data,S,K,r,T):
  opt_value=data["lastPrice"].iloc[0]
  objective_function = lambda sigma: (BS_call(S,K,r,sigma,T)- opt_value) ** 2
  result = minimize_scalar(objective_function, method="brent")

  return result.x if result.success else np.nan

sigma_pred= impl_vol_cal(call_op,S, call_op["strike"].iloc[0],0.0485,first_time)
sigma_pred
```

```python
first_opt_value=BS_call(S, call_op["strike"].iloc[0],0.0485,sigma_pred, first_time)
first_opt_value
```

## Limitations
1. The risk-free interest rate is not constant over the time and it fluctuates over the time;
2. it assumes no transaction costs and no dividends, but it is not correct in the real world scenario;
3. it is applied on the European options;
4. the model does not take into account the price swings and in the specific some stylized facts like the leverage effect and the volatility clustering;
5. the volatility is assumed to be constant, but actually it does not match the reality: if we try to plot the implied volatility against the strike prices we end up with a phenomena called "volatility smile". This is caused by the noise of the market.
We specify that the implied volatility is the volatility that plugged into the BSM formula is able to minimize the difference between the price of the option and the BSM value.

###Volatility smile

The volatility smile is a graphical pattern often observed in the implied volatilities of options. When plotted against strike prices for a specific expiration, implied volatilities create a U-shaped curve, with higher implied volatilities for both deep in-the-money and out-of-the-money options, while at-the-money options have lower implied volatility.

There are two types of Skews:
1. Forward Skew (Positive Skew): When calls have higher implied volatilities than puts. This may occur in assets expected to increase, with upward movement volatility priced in.

  A market with forward skew often reflects a positive market sentiment, with investors anticipating potential upside. This increased optimism drives demand for OTM call options, pushing their prices higher.

  Forward skew is often accompanied by a volatility skew, where OTM call options have higher implied volatility than OTM put options. Higher implied volatility translates to higher option prices, further contributing to the elevated cost of OTM calls.

  As options approach expiration, their time value decays. However, the rate of decay can differ between calls and puts, especially in markets with interest rate differentials. In some cases, the time decay of OTM calls may be slower than that of OTM puts, leading to higher prices for OTM calls.
  Market frictions, such as bid-ask spreads and transaction costs, can impact option prices. If OTM call options are more liquid than OTM put options, their prices may be less affected by these frictions, leading to a relative price advantage.

2. Reverse Skew (Negative Skew): When puts have higher implied volatilities than calls. This is common in equity options due to the “fear of a crash,” as investors often hedge against downward moves, pushing up the price (and implied volatility) of out-of-the-money puts.

  In a reverse skew, out-of-the-money (OTM) puts tend to show higher prices due to increased demand for downside protection. Since the option's intrinsic value, time to expiration, and other factors (like the underlying price) remain constant for a given set of market conditions, the only way to account for the higher price of these OTM puts is through an increase in their implied volatility (IV).
